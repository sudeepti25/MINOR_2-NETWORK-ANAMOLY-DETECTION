```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model, Model
```

```python
autoencoder = load_model("../MODELS/autoencoder.h5", compile=False)
```

```python
df = pd.read_csv("../PREPROCESSING/processed_cicids2017.csv")
```

```python
X = df.drop(columns=['Label']).values
y = df['Label'].values
```

```python
encoder = Model(
    inputs=autoencoder.input,
    outputs=autoencoder.layers[3].output
)
```

```python
encoder.summary()
```

```python
encoded_features = encoder.predict(X)
```

```python
print(encoded_features.shape)
```

```python
encoded_df = pd.DataFrame(
    encoded_features,
    columns=["enc1", "enc2", "enc3", "enc4", "enc5"]
)

encoded_df["Label"] = y
```

```python
encoded_df.to_csv("encoded_network_features.csv", index=False)
```
